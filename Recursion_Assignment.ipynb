```python
// A function that moves n disks from rod A to rod C, using rod B as an auxiliary
static void towerOfHanoi(int n, char A, char B, char C) {
  // Base case: if there is only one disk, move it from A to C and print the move
  if (n == 1) {
    System.out.println("Move disk 1 from rod " + A + " to rod " + C);
    return;
  }
  // Recursive case: move the top n-1 disks from A to B, using C as an auxiliary
  towerOfHanoi(n-1, A, C, B);
  // Move the remaining disk from A to C and print the move
  System.out.println("Move disk " + n + " from rod " + A + " to rod " + C);
  // Move the n-1 disks from B to C, using A as an auxiliary
  towerOfHanoi(n-1, B, A, C);
}
```

Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1
to word2.
Example 1:
Input: word1 = "horse", word2 = "ros"
Output: 3
Explanation:
horse -> rorse (replace 'h' with 'r')
rorse -> rose (remove 'r')
rose -> ros (remove 'e')

```python
# A function that returns the minimum edit distance between two strings
def edit_distance(word1, word2):
  # Get the lengths of the strings
  m = len(word1)
  n = len(word2)
  # Create a matrix of size (m+1) x (n+1) to store the results
  matrix = [[0 for j in range(n+1)] for i in range(m+1)]
  # Initialize the first row and column of the matrix
  for i in range(m+1):
    matrix[i][0] = i # The cost of deleting i characters from word1
  for j in range(n+1):
    matrix[0][j] = j # The cost of inserting j characters to word1
  # Fill the rest of the matrix using the recurrence relation
  for i in range(1, m+1):
    for j in range(1, n+1):
      # If the last characters are the same, no operation is needed
      if word1[i-1] == word2[j-1]:
        matrix[i][j] = matrix[i-1][j-1]
      # If the last characters are different, choose the minimum of three operations
      else:
        matrix[i][j] = 1 + min(matrix[i-1][j], # Deletion
                               matrix[i][j-1], # Insertion
                               matrix[i-1][j-1]) # Substitution
  # Return the bottom-right corner of the matrix, which is the final answer
  return matrix[m][n]
```

Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].

```python
# A function that returns the max value of an array
def max_value(array):
  # Initialize the max value to the first element of the array
  max_value = array[0]
  # Loop through the rest of the array
  for i in range(1, len(array)):
    # If the current element is larger than the max value, update the max value
    if array[i] > max_value:
      max_value = array[i]
  # Return the max value
  return max_value
# The array given in the question
array = [ 13, 1, -3, 22, 5]
# Print the max value of the array
print(max_value(array)) # Output: 22
```

Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].

```python
# A function that returns the sum of the values of an array
def sum_values(array):
  # Initialize the sum to zero
  sum = 0
  # Loop through the array
  for element in array:
    # Add the current element to the sum
    sum += element
  # Return the sum
  return sum
# The array given in the question
array = [92, 23, 15, -20, 10]
# Print the sum of the values of the array
print(sum_values(array)) # Output: 120
```

Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum
of every digit in that number raised to the power of total digits in that number is equal to the number.
Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong number. (Easy)
Input1 : 153
Output1 : Yes
Input 2 : 134
Output2 : No

```python
# A function that checks if a number is an Armstrong number or not
def is_armstrong(n):
  # Count the number of digits in n
  order = len(str(n))
  # Initialize sum to zero
  sum = 0
  # Loop through each digit of n
  for digit in str(n):
    # Raise the digit to the power of order and add it to sum
    sum += int(digit) ** order
  # Compare sum with n and return yes or no
  if sum == n:
    return "Yes"
  else:
    return "No"
# Input 1
n = 153
print(is_armstrong(n)) # Output: Yes

# Input 2
n = 134
print(is_armstrong(n)) # Output: No
```
